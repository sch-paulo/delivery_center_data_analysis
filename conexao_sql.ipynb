# Importação das bibliotecas

```python
from sqlalchemy import create_engine
import pandas as pd
import os
import glob


pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

# Definição dos parâmetros de conexão com o MySQL

```python
user = "root"   
senha = "2020"
banco_de_dados = "delivery_center"
host = "localhost"
porta = 3306
    
conexao = create_engine(url=f"mysql+pymysql://{user}:{senha}@{host}:{porta}/{banco_de_dados}")
```

# Importação dos dados

```python
# Obtém o diretório da pasta com os datasets (considerando que está na mesma pasta que o Notebook)
caminho_pasta = os.getcwd() + '\delivery_center'

# Obtém o diretório para a sua pasta contendo os arquivos CSV
caminho_pasta = os.path.join(caminho_pasta, "*.csv")

# Função glob para obter a lista de arquivos CSV na pasta
lista_arquivos = glob.glob(caminho_pasta)

lista_dfs = []

# Iteração sobre a lista de arquivos para ler cada arquivo CSV
for arquivo in lista_arquivos:
    nome_dataframe = arquivo.split("\\")[-1].split('.')[0]
    
    # Especifica a codificação do arquivo, para não haver problemas na leitura
    globals()[nome_dataframe] = pd.read_csv(arquivo, encoding='latin1')
    lista_dfs.append(nome_dataframe)
    print(nome_dataframe)
```

```python
# Amostra de cada dataframe, para checar se está tudo de acordo
for nome in lista_dfs:
    print(nome)
    display(globals()[nome].sample(10))
```

## Consertando formato de data e hora da tabela *orders*

```python
# Lista de colunas que contêm datas e horas
date_columns = ['order_moment_created', 'order_moment_accepted', 'order_moment_ready', 
                'order_moment_collected', 'order_moment_in_expedition', 'order_moment_delivering', 
                'order_moment_delivered', 'order_moment_finished']

# Função para converter o formato da data e hora
def convert_datetime(date_str):
    if date_str:
        return pd.to_datetime(date_str, format='%m/%d/%Y %I:%M:%S %p')
    else:
        return None

# Aplicar a função de conversão a todas as colunas de data e hora
for col in date_columns:
    orders[col] = orders[col].apply(convert_datetime)

orders.sample()
```

### Conexão ao MySQL

```python
for nome in lista_dfs:
    print(f'Conectando: {nome}')
    globals()[nome].to_sql(nome, conexao, index=False, if_exists='append')
    print(f'Conectado!')
```
